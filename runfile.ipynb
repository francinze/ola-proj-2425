```python
from base_classes.setting import Setting
from base_classes.environment import Environment
import matplotlib.pyplot as plt
```

```python
import numpy as np


setting = Setting(T=100)
seed = 42
np.random.seed(seed)
env = Environment(setting=setting, distribution='uniform')
```

```python
env.play_all_rounds()
```

```python
plt.plot(env.seller_utility)
plt.xlabel('Round')
plt.ylabel('Seller Utility')
plt.title('Seller Utility Over Rounds')
plt.show()
```

```python
plt.plot(env.purchases)
plt.xlabel('Round')
plt.ylabel('Purchases')
plt.title('Purchases Over Rounds')
plt.show()
```
